```python
import ipywidgets as widgets
import random

class RockPaperScissors:

  def __init__(self):

    #variables to store player and computer moves, player and computer wins and draws
    self.playerMove = -1
    self.aiMove = -1
    self.playerWins = 0
    self.aiWins = 0
    self.draws = 0

    #create the buttons for each move
    self.rockBtn = widgets.Button(description="Rock", layout=widgets.Layout(width='100px', height='100px',font_size='50px'))
    self.paperBtn = widgets.Button(description="Paper", layout=widgets.Layout(width='100px', height='100px',font_size='50px'))
    self.scissorsBtn = widgets.Button(description="Scissors", layout=widgets.Layout(width='100px', height='100px',font_size='50px'))


    #create the image variable for each move and for the player's move
    self.rockImg = open('images/rock.png', "rb").read()
    self.scisImg = open('images/scissors.png', 'rb').read()
    self.papeImg = open('images/paper.png', 'rb').read()
    self.qstnImg = open('images/question.png', 'rb').read()
    self.nothing = open('images/nothing.png', 'rb').read()
    self.wonImg = open('images/win.png', 'rb').read()
    self.lostImg = open('images/loss.png', 'rb').read()
    self.drawImg = open('images/draw.png', 'rb').read()
    self.playerImg = widgets.Image(
        value=self.qstnImg,
        format='png',
        width=310,
        height=310
    )
    self.aiImg = widgets.Image(
        value=self.qstnImg,
        format='png',
        width=310,
        height=310
    )
    self.winImg = widgets.Image(
        value = self.nothing,
        format = 'png',
        width = 415,
        height = 415
    )

    #create player wins, ai wins, draws labels
    self.playerWinsLabel = widgets.Label(value=f"Player wins: {self.playerWins}", layout = widgets.Layout(height='100px', width = '100px'))
    self.aiWinsLabel = widgets.Label(value=f"AI wins: {self.aiWins}", layout = widgets.Layout(height='100px', width = '100px'))
    self.drawsLabel = widgets.Label(value=f"Draws: {self.draws}", layout = widgets.Layout(height='100px', width = '100px'))

    #create play again button
    self.playAgainBtn = widgets.Button(description='Play again?', button_style='info',layout=widgets.Layout(width='auto', height='100px'), disabled = True)

    #create the 'confirm' button
    self.confirmBtn = widgets.Button(description="Confirm", button_style='success',layout=widgets.Layout(width='auto',height='100px'),disabled=True)

    #add button press functionality
    self.rockBtn.on_click(self.clicked)
    self.paperBtn.on_click(self.clicked)
    self.scissorsBtn.on_click(self.clicked)
    self.confirmBtn.on_click(self.clicked)
    self.playAgainBtn.on_click(self.clicked)

  def clicked(self, btn):
    match btn.description:
      case 'Rock':
        self.playerImg.value = self.rockImg
        self.playerMove = 0
        self.confirmBtn.disabled = False
      case 'Paper':
        self.playerImg.value = self.papeImg
        self.playerMove = 1
        self.confirmBtn.disabled = False
      case 'Scissors':
        self.playerImg.value = self.scisImg
        self.playerMove = 2
        self.confirmBtn.disabled = False
      case 'Confirm':
        self.rockBtn.disabled = True
        self.paperBtn.disabled = True
        self.scissorsBtn.disabled = True
        self.confirmBtn.disabled = True
        self.playAgainBtn.disabled = False
        self.aiMove = random.randint(0,2)
        self.doEndThings(self.compare())
        self.playerWinsLabel.value = f"Player wins: {self.playerWins}"
        self.aiWinsLabel.value = f"AI wins: {self.aiWins}"
        self.drawsLabel.value = f"Draws: {self.draws}"
      case 'Play again?':
        self.rockBtn.disabled = False
        self.paperBtn.disabled = False
        self.scissorsBtn.disabled = False
        self.confirmBtn.disabled = True
        self.playAgainBtn.disabled = True
        self.playerImg.value = self.qstnImg
        self.aiImg.value = self.qstnImg
        self.winImg.value = self.nothing


      case default:
        print(btn.description)


  def displayGame(self):
    #create, populate, and display the grid which contains all game elements
    self.grid = widgets.GridspecLayout(5,12,height='525px',width='1260px')
    self.grid[0,0] = self.rockBtn
    self.grid[1,0] = self.paperBtn
    self.grid[2,0] = self.scissorsBtn
    self.grid[3,0:8] = self.confirmBtn
    self.grid[4,0:8] = self.playAgainBtn
    self.grid[0:3,1:4] = self.playerImg
    self.grid[0:3,5:8] = self.aiImg
    self.grid[4,8] = self.playerWinsLabel
    self.grid[4,9] = self.aiWinsLabel
    self.grid[4,10] = self.drawsLabel
    self.grid[0:4,8:12] = self.winImg
    self.grid[1,4] = widgets.Image(value=open('images/vs.png', "rb").read(),format='png',width=100,height=100)
    display(self.grid)

  def runGame(self):
    self.displayGame()

  def compare(self): # determine ai's move and who wins
    self.aiMove = random.randint(0,2)
    match self.aiMove:
      case 0:
        self.aiImg.value = self.rockImg
      case 1:
        self.aiImg.value = self.papeImg
      case 2:
        self.aiImg.value = self.scisImg

    if (self.playerMove == 0 and self.aiMove == 2) or (self.playerMove == 1 and self.aiMove == 0) or (self.playerMove == 2 and self.aiMove == 1):
      return 'win'
    elif (self.playerMove == 0 and self.aiMove == 1) or (self.playerMove == 1 and self.aiMove == 2) or (self.playerMove == 2 and self.aiMove == 0):
      return 'loss'
    else:
      return 'draw'

  def doEndThings(self, condition): #do things depending on the end case
    self.playAgainBtn.disabled = False
    if condition == 'win':
      self.playerWins += 1
      self.winImg.value = self.wonImg
    elif condition == 'loss':
      self.aiWins += 1
      self.winImg.value = self.lostImg
    else:
      self.draws += 1
      self.winImg.value = self.drawImg

game = RockPaperScissors()
game.runGame()
```
